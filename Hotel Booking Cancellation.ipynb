# Data preparation

```python
y = df3.is_canceled 
```

```python
y.shape
```

```python
x = df3[['hotel','lead_time','stays_in_weekend_nights','stays_in_week_nights','adults','children','babies','country',
         'market_segment','is_repeated_guest','previous_cancellations','previous_bookings_not_canceled',
         'reserved_room_type','deposit_type','adr','booking_changes','agent']]
```

```python
x.head()
```

```python
x1 = pd.get_dummies(df3[['hotel','country','market_segment','reserved_room_type','deposit_type']],drop_first=True)
```

```python
x1.head()
```

```python
z = df3[['lead_time','stays_in_weekend_nights','stays_in_week_nights',
                       'adults','children','babies','is_repeated_guest',
                       'previous_cancellations','previous_bookings_not_canceled','booking_changes','adr','agent']]
```

```python
x2 = pd.concat([x1,z],axis='columns')
```

```python
x2.shape
```

```python
x2.head()
```

# Splitting test and train dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x2,y,train_size=0.7,random_state=10)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='newton-cg')
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(lr.predict(x_test),y_test)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(rf.predict(x_test),y_test)
```
